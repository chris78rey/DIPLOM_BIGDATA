```python
#hazme un ejemplo de jinga de como generar un pdf de una factura  con los datos de un diccionario

from jinja2 import Template
from weasyprint import HTML, CSS

# Datos de la factura
factura = {
    'nombre': 'Juan',
    'apellido': 'Pérez',
    'direccion': 'Calle Falsa 123',
    'ciudad': 'Madrid',
    'provincia': 'Madrid',
    'codigo_postal': '28001',
    'fecha': '01/01/2019',
    'conceptos': [
        {'descripcion': 'Concepto 1', 'cantidad': 1, 'precio': 10},
        {'descripcion': 'Concepto 2', 'cantidad': 2, 'precio': 20},
        {'descripcion': 'Concepto 3', 'cantidad': 3, 'precio': 30},
    ],
    'iva': 21,
    'forma_pago': 'Transferencia',
}

# Plantilla HTML
plantilla = Template("""
<!DOCTYPE html>
<html>

<head>

    <meta charset="utf-8">
    <title>Factura</title>
    <style>
        body {
            font-family: Arial, Helvetica, sans-serif;  
        }   
        table { 
            border-collapse: collapse;  
        }   
        table, th, td { 
            border: 1px solid black;    
        }   
        th, td {    
            padding: 5px;   
        }   
        th {    
            text-align: left;   
        }   
    </style>
</head> 

<body>
    <h1>Factura</h1>
    <p>Nombre: {{ nombre }} {{ apellido }}</p>  
    <p>Dirección: {{ direccion }}</p>   
    <p>Ciudad: {{ ciudad }}</p> 
    <p>Provincia: {{ provincia }}</p>   
    <p>Código postal: {{ codigo_postal }}</p>   
    <p>Fecha: {{ fecha }}</p>   
    <table style="width:100%">  
        <tr>
            <th>Concepto</th>
            <th>Cantidad</th>   
            <th>Precio</th> 
            <th>Subtotal</th>   
        </tr>   
        {% for concepto in conceptos %} 
        <tr>    
            <td>{{ concepto.descripcion }}</td>
            <td>{{ concepto.cantidad }}</td>    
            <td>{{ concepto.precio }}</td>  
            <td>{{ concepto.cantidad * concepto.precio }}</td>  
        </tr>   
        {% endfor %}    
        <tr>    
            <td colspan="3">Subtotal</td>
            <td>{{ subtotal }}</td> 
        </tr>   
        <tr>    
            <td colspan="3">IVA ({{ iva }}%)</td>
            <td>{{ iva_total }}</td>    
        </tr>   
        <tr>    
            <td colspan="3">Total</td>
            <td>{{ total }}</td>    
        </tr>   
    </table>    
    <p>Forma de pago: {{ forma_pago }}</p>  
</body> 

</html> 
""")    

# Generar HTML  
html = plantilla.render(**factura)  

# Generar PDF   
HTML(string=html).write_pdf('factura.pdf', stylesheets=[CSS(string='@page { size: A4; margin: 1cm }')]) 

# Mostrar PDF   
from IPython.display import IFrame  
IFrame('factura.pdf', width=600, height=300)    

# Mostrar HTML
from IPython.display import HTML    
HTML(html)  

# Mostrar HTML en una nueva ventana 
import webbrowser   
webbrowser.open('factura.html', new=2)  

# Guardar HTML  
with open('factura.html', 'w') as f:    
    f.write(html)   

# Guardar PDF   
HTML(string=html).write_pdf('factura.pdf', stylesheets=[CSS(string='@page { size: A4; margin: 1cm }')]) 

# Mostrar PDF   
from IPython.display import IFrame  
IFrame('factura.pdf', width=600, height=300)    

# Mostrar HTML  
from IPython.display import HTML    
HTML(html)  
```
