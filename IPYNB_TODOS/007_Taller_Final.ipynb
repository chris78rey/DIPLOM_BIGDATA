```python
icfes = pd.read_csv('https://raw.githubusercontent.com/HarryVargas96/ICFES/main/icfes_2020_reducido.csv')
```

```python
'''gfuardar el dataset icfes en un archivo llamado icfes.csv'''
icfes.to_csv('icfes.csv', index=False)

```
